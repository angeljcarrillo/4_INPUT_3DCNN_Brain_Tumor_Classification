```python
import os
import numpy as np
import tensorflow as tf
import nibabel as nib
from scipy import ndimage
from tensorflow import keras
from tensorflow.keras import layers
import re
import matplotlib.pyplot as plt
import matplotlib as mpl
import random
from utils import remove_store, get_data, read_nifti_file, resize_volume, process_scan
import keras_tuner as kt
```

```python
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
```

```python
low_grade_t1_p = get_data(tumor='LGG', pattern='t1')
high_grade_t1_p = get_data(tumor='HGG', pattern='t1')
low_grade_t1ce_p = get_data(tumor='LGG', pattern='t1ce')
high_grade_t1ce_p = get_data(tumor='HGG', pattern='t1ce')
low_grade_flair_p = get_data(tumor='LGG', pattern='flair')
high_grade_flair_p = get_data(tumor='HGG', pattern='flair')
low_grade_t2_p = get_data(tumor='LGG', pattern='t2')
high_grade_t2_p = get_data(tumor='HGG', pattern='t2')
```

```python
print(len(low_grade_t1_p), len(low_grade_t1ce_p), len(low_grade_flair_p), len(low_grade_t2_p))
print(len(high_grade_t1_p), len(high_grade_t1ce_p), len(high_grade_flair_p), len(high_grade_t2_p))
```

```python
# Read and process the scans.
# Each scan is resized across height, width, and depth and rescaled.
low_grade_t1 = np.array([process_scan(path) for path in low_grade_t1_p])
high_grade_t1 = np.array([process_scan(path) for path in high_grade_t1_p])
low_grade_t1ce = np.array([process_scan(path) for path in low_grade_t1ce_p])
high_grade_t1ce = np.array([process_scan(path) for path in high_grade_t1ce_p])
low_grade_flair = np.array([process_scan(path) for path in low_grade_flair_p])
high_grade_flair = np.array([process_scan(path) for path in high_grade_flair_p])
low_grade_t2 = np.array([process_scan(path) for path in low_grade_t2_p])
high_grade_t2 = np.array([process_scan(path) for path in high_grade_t2_p])
```

```python
# For the CT scans having presence of viral pneumonia
# assign 1, for the normal ones assign 0.
low_grade_labels = np.array([1 for _ in range(len(low_grade_t1))])
high_grade_labels = np.array([0 for _ in range(len(high_grade_t1))])
```

```python
train_low = int(len(low_grade_labels) * 0.7)
train_high = int(len(high_grade_labels) * 0.7)
# Split data in the ratio 70-30 for training and validation.
t1_train = np.concatenate((low_grade_t1[:train_low], high_grade_t1[:train_high]), axis=0)
t1_val = np.concatenate((low_grade_t1[train_low:], high_grade_t1[train_high:]), axis=0)
del low_grade_t1, high_grade_t1
t1ce_train = np.concatenate((low_grade_t1ce[:train_low], high_grade_t1ce[:train_high]), axis=0)
t1ce_val = np.concatenate((low_grade_t1ce[train_low:], high_grade_t1ce[train_high:]), axis=0)
del low_grade_t1ce, high_grade_t1ce
flair_train = np.concatenate((low_grade_flair[:train_low], high_grade_flair[:train_high]), axis=0)
flair_val = np.concatenate((low_grade_flair[train_low:], high_grade_flair[train_high:]), axis=0)
del low_grade_flair, high_grade_flair
t2_train = np.concatenate((low_grade_t2[:train_low], high_grade_t2[:train_high]), axis=0)
t2_val = np.concatenate((low_grade_t2[train_low:], high_grade_t2[train_high:]), axis=0)
del low_grade_t2, high_grade_t2
y_train = np.concatenate((low_grade_labels[:train_low], high_grade_labels[:train_high]), axis=0)
y_val = np.concatenate((low_grade_labels[train_low:], high_grade_labels[train_high:]), axis=0)
```

```python
def get_model(width=128, height=128, depth=64):
    """Build a 3D convolutional neural network model."""

    inputs_1 = keras.Input((width, height, depth, 1))
    inputs_2 = keras.Input((width, height, depth, 1))
    inputs_3 = keras.Input((width, height, depth, 1))
    inputs_4 = keras.Input((width, height, depth, 1))
    
    # First cnn

    x = layers.Conv3D(filters=64, kernel_size=3, activation="relu")(inputs_1)
    x = layers.MaxPool3D(pool_size=2)(x)
    x = layers.BatchNormalization()(x)

    x = layers.Conv3D(filters=64, kernel_size=3, activation="relu")(x)
    x = layers.MaxPool3D(pool_size=2)(x)
    x = layers.BatchNormalization()(x)

    x = layers.Conv3D(filters=128, kernel_size=3, activation="relu")(x)
    x = layers.MaxPool3D(pool_size=2)(x)
    x = layers.BatchNormalization()(x)

    x = layers.Conv3D(filters=256, kernel_size=3, activation="relu")(x)
    x = layers.MaxPool3D(pool_size=2)(x)
    x = layers.BatchNormalization()(x)

    x = layers.GlobalAveragePooling3D()(x)
    
    model_1 = keras.Model(inputs_1, x, name="t1_cnn")
    
    ### Second cnn
    
    y = layers.Conv3D(filters=64, kernel_size=3, activation="relu")(inputs_2)
    y = layers.MaxPool3D(pool_size=2)(y)
    y = layers.BatchNormalization()(y)

    y = layers.Conv3D(filters=64, kernel_size=3, activation="relu")(y)
    y = layers.MaxPool3D(pool_size=2)(y)
    y = layers.BatchNormalization()(y)

    y = layers.Conv3D(filters=128, kernel_size=3, activation="relu")(y)
    y = layers.MaxPool3D(pool_size=2)(y)
    y = layers.BatchNormalization()(y)

    y = layers.Conv3D(filters=256, kernel_size=3, activation="relu")(y)
    y = layers.MaxPool3D(pool_size=2)(y)
    y = layers.BatchNormalization()(y)

    y = layers.GlobalAveragePooling3D()(y)
    
    model_2 = keras.Model(inputs_2, y, name="t1ce_cnn")
    
    ### Third cnn
    
    w = layers.Conv3D(filters=64, kernel_size=3, activation="relu")(inputs_3)
    w = layers.MaxPool3D(pool_size=2)(w)
    w = layers.BatchNormalization()(w)

    w = layers.Conv3D(filters=64, kernel_size=3, activation="relu")(w)
    w = layers.MaxPool3D(pool_size=2)(w)
    w = layers.BatchNormalization()(w)

    w = layers.Conv3D(filters=128, kernel_size=3, activation="relu")(w)
    w = layers.MaxPool3D(pool_size=2)(w)
    w = layers.BatchNormalization()(w)

    w = layers.Conv3D(filters=256, kernel_size=3, activation="relu")(w)
    w = layers.MaxPool3D(pool_size=2)(w)
    w = layers.BatchNormalization()(w)

    w = layers.GlobalAveragePooling3D()(w)
    
    model_3 = keras.Model(inputs_3, w, name="flair")
    
    ## Last cnn
    
    t = layers.Conv3D(filters=64, kernel_size=3, activation="relu")(inputs_4)
    t = layers.MaxPool3D(pool_size=2)(t)
    t = layers.BatchNormalization()(t)

    t = layers.Conv3D(filters=64, kernel_size=3, activation="relu")(t)
    t = layers.MaxPool3D(pool_size=2)(t)
    t = layers.BatchNormalization()(t)

    t = layers.Conv3D(filters=128, kernel_size=3, activation="relu")(t)
    t = layers.MaxPool3D(pool_size=2)(t)
    t = layers.BatchNormalization()(t)

    t = layers.Conv3D(filters=256, kernel_size=3, activation="relu")(t)
    t = layers.MaxPool3D(pool_size=2)(t)
    t = layers.BatchNormalization()(t)

    t = layers.GlobalAveragePooling3D()(t)
    
    model_4 = keras.Model(inputs_4, t, name="t2")
    
    ## Dense layer
    
    combined = layers.concatenate([model_1.output, model_2.output, model_3.output, model_4.output])
    
    z = layers.Dense(units=512, activation="relu")(combined)
    z = layers.Dense(units=256, activation="relu")(z)
    z = layers.Dropout(0.3)(z)

    outputs = layers.Dense(units=1, activation="sigmoid")(z)

    # Define the model.
    model = keras.Model(inputs=[inputs_1, inputs_2, inputs_3, inputs_4], outputs=outputs)
    return model


# Build model.
model = get_model(width=128, height=128, depth=64)
#model.summary()
```

```python
# Train model
# Compile model.
initial_learning_rate = 0.0001
lr_schedule = keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate, decay_steps=100000, decay_rate=0.96, staircase=True
)
model.compile(
    loss="binary_crossentropy",
    optimizer=keras.optimizers.Adam(learning_rate=lr_schedule),
    metrics=["acc"],
)

# Define callbacks.
checkpoint_cb = keras.callbacks.ModelCheckpoint(
    "../neural_summary/3d_image_classification.h5", save_best_only=True
)
early_stopping_cb = keras.callbacks.EarlyStopping(
    monitor="val_acc", patience=15)

# Train the model, doing validation at the end of each epoch
epochs = 50
model.fit(
    x=[t1_train, t1ce_train, flair_train, t2_train],
    y=y_train,
    batch_size=2,
    validation_data=([t1_val, t1ce_val, flair_val, t2_val], y_val),
    epochs=epochs,
    shuffle=True,
    verbose=2,
    callbacks=[checkpoint_cb, early_stopping_cb],
)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 3))
ax = ax.ravel()

for i, metric in enumerate(["acc", "loss"]):
    ax[i].plot(model.history.history[metric])
    ax[i].plot(model.history.history["val_" + metric])
    ax[i].set_title("Model {}".format(metric))
    ax[i].set_xlabel("epochs")
    ax[i].set_ylabel(metric)
    ax[i].legend(["train", "val"])
fig.savefig('../neural_summary/net_performance_multi.jpg')
```
