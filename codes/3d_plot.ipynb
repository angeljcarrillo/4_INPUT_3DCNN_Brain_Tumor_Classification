```python
import matplotlib.pyplot as plt
import numpy as np
import nibabel as nib
```

```python
def read_nifti_file(filepath):
    """Read and load volume"""
    # Read file
    scan = nib.load(filepath)
    # Get raw data
    scan = scan.get_fdata()
    return scan
```

```python
image = read_nifti_file('../data/LGG/BraTS20_Training_260_flair.nii.gz')
```
